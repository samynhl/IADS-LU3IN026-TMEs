```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
x1 = np.random.randint(9, size=(20, 1))
x2 = np.random.randint(9, size=(20, 1))
```

```python
plt.scatter(x1,x2)
plt.plot([0,1,9],[0,1,9],"r")
```

```python
plt.scatter(x1,x2)
m=1
b=0
plt.plot(x1, m*x1 + b, "r")
```

```python
plt.plot(x1, m*x1 + b, "r")
```

```python
col = np.where(m*x1 + b<x2,'r')

plt.scatter(x1, x2, c=col, s=5)
plt.show()
```
