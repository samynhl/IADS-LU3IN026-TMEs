```python
x = np.linspace(0,10,100)
y = x**2
z = x*2
y = np.sqrt(x)
plt.plot(x,y,'r',label='puissance')
plt.plot(x,z,'g', label='multiplication')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("X vs Y")
plt.show()
```

```python
# Now if you want to plot mutliple plot together you can use subplot()
plt.subplot(1,2,1)
plt.plot(x,y,'b')
# subplot(no of rows, no of columns, plot no you are refering to)
plt.subplot(1,2,2)
plt.plot(x,y,'r')
plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
np.random.seed(19839830)
N = 40

x = np.random.randint(20, size=N)
y = np.random.randint(20, size=N)

col = np.where(x<0,'k',np.where(y<(-1.09375*x+17.5),'b','r'))

plt.scatter(x, y, c=col, s=10)

plt.plot(x, -1.09375*x+17.5, c='g')
```
